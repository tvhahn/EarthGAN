```python
import xarray as xr
import pathlib
from pathlib import Path
import numpy as np
# import pyvista as pv
import matplotlib.pyplot as plt
import torch
```

```python
# make your root_dir the current working directory
root_dir = Path.cwd().parent.parent # set the root directory as a Pathlib path
print(root_dir)

path_processed_data = root_dir / 'data/processed/' # raw data folder that holds the .zip .mat files for milling data
path_truth = path_processed_data / "truth"
```

```python
# get downsampled data based on time setp
truth_name = f'x_truth_001.npy'
x = torch.tensor(np.load(path_truth / truth_name))
```

```python
# https://stackoverflow.com/a/34594651/9214620
# https://stackoverflow.com/a/46069458/9214620

ax = plt.subplot(111, projection='mollweide')
# fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18,8), subplot_kw={'projection': ccrs.Mollweide()})
# ax
# ax.subplot(111, projection='mollweide')
lon = np.linspace(-np.pi, np.pi,216)
lat = np.linspace(-np.pi/2., np.pi/2.,108)
Lon,Lat = np.meshgrid(lon,lat)
ax.pcolormesh(Lon,Lat, x[0,1,3,:,:], cmap='inferno')
# ax.subplot(111, projection='mollweide')
plt.show()
```
